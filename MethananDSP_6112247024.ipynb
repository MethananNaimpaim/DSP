```python
#ประกาศคำสั่ง
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot

import numpy as np

from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display
```

```python
#lab01
sin_sig = thinkdsp.SinSignal(freq=7024, amp=24, offset=0)

sin_sig.plot()
thinkplot.config(xlabel='Time (s)')
```

```python
#lab02
signal = thinkdsp.TriangleSignal(240)                                     #ความถี่ของเสียง้ทุ่้มแหลม
duration = signal.period*400                                               #ช่วงเวลาที่เสียงดัง
segment = signal.make_wave(duration,framerate =10000)
segment.plot()

MNwave = signal.make_wave(duration,framerate=10000)

spectrum =  MNwave.make_spectrum()
spectrum.plot()
MNwave.apodize()
MNwave.make_audio()
```

```python

signal = thinkdsp.SquareSignal(4000)                                     #ความถี่ของเสียง้ทุ่้มแหลม
duration = signal.period*4500                                               #ช่วงเวลาที่เสียงดัง
segment = signal.make_wave(duration,framerate =10000)
segment.plot()

Mwave = signal.make_wave(duration,framerate=10000)

Mwave.apodize()
Mwave.make_audio()
```
